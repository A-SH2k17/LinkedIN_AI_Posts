## script for financial times

```python
main_url = 'https://www.ft.com/technology?page='

extracted_articles = {}

for i in range(1,2):
    result = requests.get(main_url + str(i))
    soup = BeautifulSoup(result.content, 'html.parser')

    article_section = soup.select_one('ul[data-stream-list=""]')
    article_list = article_section.select('li.o-teaser-collection__item')
    for article in article_list:
        title = article.select_one('a.js-teaser-heading-link')
        if not title:
            continue
        link = article.select_one('a.js-teaser-heading-link')['href']
        summary = article.select_one('p.o-teaser__standfirst')
        image_link = article.select_one('img.o-teaser__image')["data-src"] if article.select_one('img.o-teaser__image') else None
        extracted_articles[title.text.strip()] = {
            'link': link,
            'summary': summary.text.strip() if summary else "No summary available",
            'image_link': image_link
        }

extracted_articles
```

```python
from playwright.sync_api import sync_playwright
from bs4 import BeautifulSoup

URL = "https://www.cnbc.com/finance/"
TARGET_ARTICLE_COUNT = 100   # <-- change this to whatever number you want


def extract_articles(html):
    """Parse current page HTML and pull out title/link/image for each
    article card inside the 'More In Finance' ThreeUpStack sections."""
    soup = BeautifulSoup(html, "html.parser")

    more_in_finance_heading = None
    for header in soup.find_all("div", class_="ModuleHeader-container"):
        if "More In Finance" in header.get_text():
            more_in_finance_heading = header
            break

    if not more_in_finance_heading:
        return []

    start_section = more_in_finance_heading.find_parent("div", class_="ThreeUpStack-container")
    if not start_section:
        return []

    sections = [start_section]
    for sib in start_section.find_next_siblings("div"):
        classes = sib.get("class", [])
        if "ThreeUpStack-container" in classes:
            sections.append(sib)
        elif "LoadMore-container" in classes:
            break

    articles = []
    seen_links = set()

    for section in sections:
        for card in section.find_all("div", attrs={"data-test": "Card"}):
            title_tag = card.find("a", class_="Card-title")
            if not title_tag:
                continue

            link = title_tag.get("href")
            if link in seen_links:
                continue
            seen_links.add(link)

            img_tag = card.find("img", class_="Card-mediaContainerInner")
            articles.append({
                "title": title_tag.get_text(strip=True),
                "link": link,
                "image": img_tag.get("src") if img_tag else None,
            })

    return articles


def scrape_with_load_more(target_count):
    with sync_playwright() as p:
        browser = p.chromium.launch(headless=True)
        page = browser.new_page()
        page.goto(URL, wait_until="domcontentloaded")

        # Dismiss cookie/consent banners if they pop up (best-effort, ignore failures)
        try:
            page.click("text=Accept", timeout=3000)
        except Exception:
            pass

        articles = extract_articles(page.content())

        max_clicks = 30  # safety cap so it doesn't loop forever
        clicks = 0

        while len(articles) < target_count and clicks < max_clicks:
            load_more_btn = page.locator("button:has-text('Load More')").first

            if not load_more_btn.is_visible():
                print("No more 'Load More' button found — site has no more articles to load.")
                break

            load_more_btn.scroll_into_view_if_needed()
            load_more_btn.click()

            # Wait for new cards to actually render before re-parsing
            page.wait_for_timeout(1500)

            articles = extract_articles(page.content())
            clicks += 1
            print(f"Click #{clicks}: now have {len(articles)} articles")

        browser.close()
        return articles[:target_count]


if __name__ == "__main__":
    results = scrape_with_load_more(TARGET_ARTICLE_COUNT)

    for a in results:
        print(a)

    print(f"\nTotal articles collected: {len(results)}")
```
