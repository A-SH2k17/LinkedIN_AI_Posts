## 1. Euklidische Distanz ($L_2$-Distanz)

### 🔹 Konzeptuelle Definition
Die Euklidische Distanz misst die Ähnlichkeit zwischen Vektoren, indem sie die Stärke des **Differenz-Vektors** ($\vec{v}_1 - \vec{v}_2$) berechnet. Je geringer der Betrag (die Magnitude) dieses Differenz-Vektors ist, desto geometrisch ähnlicher sind sich die beiden Punkte.

### 🔹 Mathematische Formel
Für verallgemeinerte $n$-dimensionale Vektoren $A$ und $B$:
$$A = (a_1, a_2, \dots, a_n)$$
$$B = (b_1, b_2, \dots, b_n)$$
$$d(A,B) = \sqrt{\sum (A_i - B_i)^2}$$

### 🔹 Anwendungsregel
Nutzen Sie die Euklidische Distanz, wenn **tatsächliche Zahlen und ihre absoluten Differenzen eine entscheidende Bedeutung tragen**.

```python
# --- PRAKTISCHE BEISPIELE FÜR EUKLIDISCHE DISTANZ ---

# Beispiel A: Ähnliche Immobilien basierend auf Preis & Größe finden
target_mansion  = np.array([5, 5, 45, 1200]) 
similar_mansion = np.array([5, 4, 42, 1150]) 
small_condo     = np.array([1, 1,  8,  150])  

dist_mansion = euclidean(target_mansion, similar_mansion)
dist_condo   = euclidean(target_mansion, small_condo)

print("--- Ergebnisse Immobilien-Matching ---")
print(f"Distanz zu ähnlicher Villa: {dist_mansion:.2f} -> Interpretation: [SEHR NAH] Datenpunkte liegen dicht beieinander.")
print(f"Distanz zu kleiner Wohnung:  {dist_condo:.2f} -> Interpretation: [NICHT ÄHNLICH] Enormer struktureller Größenunterschied.\n")

# Beispiel B: K-Means Clustering (Gruppierung von Punkten)
customers = np.array([
    [21, 45],   # Junger Berufstätiger A
    [23, 48],   # Junger Berufstätiger B
    [54, 320],  # Executive A
    [52, 290]   # Executive B
])

print("--- Kundensegment-Analyse ---")
peer_gap = np.linalg.norm(customers[0] - customers[1])
tier_gap = np.linalg.norm(customers[0] - customers[2])

print(f"Distanz zwischen Kollegen im selben Segment: {peer_gap:.2f}  -> Interpretation: [SEHR NAH] Gehört in denselben Cluster.")
print(f"Distanz beim Sprung in das Executive-Segment: {tier_gap:.2f} -> Interpretation: [WEIT ENTFERNT] Gehört in separate Cluster.")
```

## 2. Skalarprodukt & Cosinus-Ähnlichkeit

### 🔹 Definition des Skalarprodukts (Dot Product)
Das Skalarprodukt misst die Ähnlichkeit zwischen Vektoren, indem es die **Projektion von Vektor $\vec{v}$ auf Vektor $\vec{u}$** berechnet. 
$$\vec{v} \cdot \vec{u} = \|\vec{v}\| \|\vec{u}\| \cos\theta$$

### 🔹 Definition der Cosinus-Ähnlichkeit
Wie groß ist die Magnitude von $\vec{v}$ in Richtung von $\vec{u}$? Der von Vektor $\vec{v}$ auf $\vec{u}$ geworfene „Schatten“ wird durch $\|\vec{v}\|\cos\theta$ ausgedrückt.

Die Cosinus-Ähnlichkeit isoliert diese Beziehung, indem sie **sich ausschließlich für die Winkel zwischen den Vektoren interessiert** und die Gesamtlänge ignoriert. Sie wird verwendet, wenn die Verhältnisse zwischen den Merkmalen unabhängig von der Vektorgröße entscheidend sind.
$$\cos\theta = \frac{\vec{v} \cdot \vec{u}}{\|\vec{v}\| \|\vec{u}\|}$$

### 🔹 Produktionsregel
* **Cosinus-Ähnlichkeit:** Der Wertebereich liegt strikt zwischen $[-1, 1]$. Hauptanwendungen sind Textklassifikation, NLP, LLM-Embeddings, semantische Suche und Collaborative Filtering.
* **Skalarprodukt:** Wird verwendet, wenn **Ausführungsgeschwindigkeit entscheidend ist**. In Produktionsdatenbanken normieren wir Vektoren vorab auf die Länge 1 ($\|\vec{v}\|=1$). Dadurch entspricht das Skalarprodukt exakt der Cosinus-Ähnlichkeit, läuft aber um ein Vielfaches schneller.

```python
# --- PRAKTISCHE BEISPIELE FÜR WINKEL-ÄHNLICHKEIT ---
user_profile = np.array([2, 0])     
short_tweet  = np.array([1, 0])     
massive_blog = np.array([50, 0])    
cooking_recipe = np.array([0, 5])   

print("--- Vektor-Vergleichs-Analyse ---")

# 1. Gleiche Richtung, kleine Skalierung
cos_short = cosine_similarity([user_profile], [short_tweet])[0][0]
print(f"📊 User Profile vs Kurzer Tweet (Gleiche Richtung):")
print(f" -> Cosinus-Ähnlichkeit: {cos_short:.4f} -> Interpretation: [PERFEKTE ÜBEREINSTIMMUNG (1.0)] Identischer Inhalt.")

# 2. Gleiche Richtung, massive Skalierungsunterschiede
cos_blog = cosine_similarity([user_profile], [massive_blog])[0][0]
print(f"📊 User Profile vs Riesiger Blog (Gleiche Richtung, massive Skalierung):")
print(f" -> Cosinus-Ähnlichkeit: {cos_blog:.4f} -> Interpretation: [PERFEKTE ÜBEREINSTIMMUNG (1.0)] Identischer Inhalt! Skalierung wird ignoriert.")

# 3. Orthogonaler Vektor
cos_cook = cosine_similarity([user_profile], [cooking_recipe])[0][0]
print(f"📊 User Profile vs Koch-Rezept (Orthogonal / 90 Grad):")
print(f" -> Cosinus-Ähnlichkeit: {cos_cook:.4f} -> Interpretation: [KEINE ÄHNLICHKEIT (0.0)] Keine thematische Überschneidung vorhanden.")
```

## 3. Schnelle Interpretationsmatrix (Quick Interpretation)

Wir reproduzieren nun programmgesteuert die exakte Wahrheitstabelle aus den handschriftlichen Notizen, um die Grenzen unserer Metriken bei verschiedenen strukturellen Vektorbeziehungen zu verdeutlichen.

```python
# Strukturelle Vektorpaare definieren (analog zur handschriftlichen Tabelle)
vec_A = np.array([3, 0])

relationships_de = {
    "Identische Vektoren (Identical)": np.array([3, 0]),
    "Gleiche Richtung, andere Längen": np.array([12, 0]),
    "Orthogonal (90°-Winkel)": np.array([0, 3]),
    "Entgegengesetzte Richtung": np.array([-3, 0])
}

print(f"{'Vektor-Beziehung':<35} | {'Skalarprodukt':<15} | {'Cosinus-Ähnlichkeit':<22} | {'Euklidische Distanz'}")
print("-" * 100)

for name, vec_B in relationships_de.items():
    dot_p = np.dot(vec_A, vec_B)
    cos_s = cosine_similarity([vec_A], [vec_B])[0][0]
    eucl_d = euclidean(vec_A, vec_B)
    
    print(f"{name:<35} | {dot_p:<15.1f} | {cos_s:<22.4f} | {eucl_d:.4f}")
```
